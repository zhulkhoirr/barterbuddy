#Generate Dataset Aplikasi Pertukaran Barang

## Import library

```python
import pandas as pd
import random
from datetime import datetime, timedelta
from collections import defaultdict
```

## Generate tanggal acak

```python
def random_date(start, end):
  return start + timedelta(seconds=random.randint(0, int((end - start).total_seconds())))
```

##Data User

```python
users_data = []
cities = [
    "Banda Aceh", "Medan", "Padang", "Pekanbaru", "Jambi", "Palembang",
    "Bengkulu", "Bandar Lampung", "Pangkal Pinang", "Tanjung Pinang",
    "Jakarta", "Bandung", "Semarang", "Yogyakarta", "Surabaya", "Serang",
    "Denpasar", "Mataram", "Kupang", "Pontianak", "Palangkaraya", "Banjarmasin",
    "Samarinda", "Tanjung Selor", "Manado", "Gorontalo", "Palu", "Makassar",
    "Mamuju", "Kendari", "Manokwari", "Jayapura", "Ambon", "Sofifi", "Sorong",
    "Wanggar", "Walesi", "Salor"
]

for i in range(1, 301):
  users_data.append([i, f'user{i}', f'pass{i}', random.choice(cities)])

users_df = pd.DataFrame(users_data, columns=['id', 'username', 'password', 'city'])
```

##Data Post

```python
posts_data = []
barang_titles = ['Sepeda', 'Laptop', 'Smartphone', 'Televisi', 'Kamera', 'Rak Buku', 'Meja', 'Kursi', 'Lemari', 'Jam Tangan']
jasa_titles = ['Jasa Cuci Mobil', 'Jasa Pindahan', 'Jasa Kebersihan', 'Les Matematika', 'Les Bahasa Inggris', 'Jasa Tukang Kebun', 'Jasa Renovasi Rumah', 'Jasa Perawatan Hewan', 'Jasa Fotografi', 'Pelajaran Gitar']
statuses = ['Tersedia', 'Selesai']
types = ['Barang', 'Jasa']
start_date = datetime.strptime('2024-05-20 10:00:00', '%Y-%m-%d %H:%M:%S')
end_date = datetime.strptime('2024-06-10 15:00:00', '%Y-%m-%d %H:%M:%S')

title_count = defaultdict(int)

for i in range (1, 301):
  if random.choice(types) == 'Barang':
    title = random.choice(barang_titles)
    type_ = 'Barang'
    description = f'Sebuah {title} bekas'
  elif random.choice(types) == 'Jasa':
    title = random.choice(jasa_titles)
    type_ = 'Jasa'
    description = f'Sebuah {title}'
  status = random.choice(statuses)
  created_at = random_date(start_date, end_date)
  updated_at = created_at + timedelta(hours=random.randint(1, 10))
  user_id = random.randint(1, 300)

  title_count[title] += 1
  image_name = f'{title.lower()}{title_count[title]}.jpg'

  posts_data.append([i, title, image_name, description, type_, status, created_at, updated_at, user_id])

posts_df = pd.DataFrame(posts_data, columns=['id', 'title', 'image', 'description', 'type', 'status', 'created_at', 'updated_at', 'user_id'])
```

##Data Transaksi

```python
transactions_data = []

selesai_posts = posts_df[posts_df['status'] == 'Selesai']

for i in range(1, 301):
    if not selesai_posts.empty:
        post = selesai_posts.sample().iloc[0]
        post_id = post['id']
        date = random_date(start_date, end_date)
        transactions_data.append([i, post_id, date])

transactions_df = pd.DataFrame(transactions_data, columns=['id', 'post_id', 'date'])
```

```python
transactions_df.sample(5)
```

```python
transactions_df.drop_duplicates(subset=['post_id'], inplace=True)
print("Jumlah duplikasi:",  transactions_df['post_id'].duplicated().sum())
```

##Data History

```python
search_history_data = []

title = posts_df['title']

for i in range(1, 301):
    user_id = random.randint(1, 300)
    search = random.choice(title)
    search_history_data.append([i, user_id, search])

search_history_df = pd.DataFrame(search_history_data, columns=['id', 'user_id', 'search'])
```

##Data Interest

```python
interests_data = []

for i in range(1, 301):
    user_id = random.randint(1, 300)
    post_id = random.randint(1, 300)
    interests_data.append([i, user_id, post_id])

interests_df = pd.DataFrame(interests_data, columns=['id', 'user_id', 'post_id'])
```

##Export ke CSV

```python
users_df.to_csv('users.csv', index=False)
posts_df.to_csv('posts.csv', index=False)
transactions_df.to_csv('transactions.csv', index=False)
search_history_df.to_csv('search_history.csv', index=False)
interests_df.to_csv('interests.csv', index=False)
```
